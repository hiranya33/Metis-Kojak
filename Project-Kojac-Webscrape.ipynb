```python
def get_img_tags(url):
    links=[]
    for i in driver.find_elements_by_xpath('//a[@class="pt_thumb_link "]/img'):
        links.append(i.get_attribute('src'))
    return links
```

```python
def get_shoe_pg(url):
    shoe_l=[]
    for i in driver.find_elements_by_xpath('//a[@class="pt_link"]'):
        shoe_l.append(i.get_attribute('href'))
    return shoe_l
```

```python
def get_shoe_price(url):
    shoe_p=[]
    for i in driver.find_elements_by_xpath('//span[@class="pt_price"]'):
        shoe_p.append(i.text)
    return shoe_p
```

```python
def get_shoe_desc(url):
    shoe_des=[]
    for i in driver.find_elements_by_xpath('//span[@class="pt_title"]'):
        shoe_des.append(i.text)
    return shoe_des
```

```python
def get_url(page):
    if page == 1:
        return f'https://www.shoes.com/womens-sneakers/category_189'
    return f'https://www.shoes.com/womens-sneakers/category_189?page={page}'
```

```python
page = 1
itags=[]
shoe_descs=[]
shoe_pages=[]
prices=[]
while page < 74:
    url = get_url(page)
    try:
        driver = webdriver.Chrome(chromedriver)
        driver.get(url)
        time.sleep(2)
        for i in range(1,11):
            driver.execute_script(f"window.scrollTo(0, document.body.scrollHeight*{i*0.1});")
            time.sleep(0.2)
    except:
        # out of pages!
        break
    #list_of_tables.append(pd.read_html(driver.page_source)[5])
    itag=get_img_tags(url)
    itags.extend((itag))
    shoe_desc=get_shoe_desc(url)
    shoe_descs.extend((shoe_desc))
    price=get_shoe_price(url)
    prices.extend((price))
    shoe_page=get_shoe_pg(url)
    shoe_pages.extend((shoe_page))
    
    page = page + 1
    driver.close()
```
